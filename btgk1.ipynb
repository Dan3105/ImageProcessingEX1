```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

#### Constrat-limited histogram equalization

```python
image_constrat_reader = cv2.imread('constrast/moon.jpg', cv2.IMREAD_GRAYSCALE)
plt.xticks([])
plt.yticks([])
plt.imshow(image_constrat_reader, cmap='gray')
```

Biểu đồ mức sáng

```python
plt.yticks([])
hist = plt.hist(image_constrat_reader.reshape(-1), 256, [0, 255])
plt.title('Biểu đồ mức sáng của hình ảnh moon trước khi áp dụng thuật toán')
plt.xlim([0, 255])
```

```python
img_hist = cv2.equalizeHist(image_constrat_reader)
res = np.hstack((image_constrat_reader, img_hist))
plt.xticks([])
plt.yticks([])
plt.imshow(res, cmap='gray')
```

```python
plt.yticks([])
hist = plt.hist(img_hist.reshape(-1), 256, [0, 255])
plt.title('biểu đồ mức sáng sau khi áp dụng thuật toán')
plt.xlim([0, 255])
```

Adaptive histogram equalization

```python
filter_size = (16,16)
dental_img = cv2.imread('adaptive/dental.jpg', cv2.IMREAD_GRAYSCALE)
parrot_img = cv2.imread('adaptive/parrot.jpg', cv2.IMREAD_GRAYSCALE)
skull_img = cv2.imread('adaptive/skull.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
clache_16 = cv2.createCLAHE(clipLimit=2.0, tileGridSize=filter_size)
adapt_16_skull = clache_16.apply(skull_img)
adapt_16_parrot = clache_16.apply(parrot_img)
adapt_16_dental = clache_16.apply(dental_img)
```

```python
plt.figure(6, figsize=(10,6))

plt.subplot(3,3,1)
plt.xticks([])
plt.yticks([])
plt.title('original_image')
plt.imshow(dental_img, cmap='gray')

plt.subplot(3,3,2)
plt.xticks([])
plt.yticks([])
plt.title('original_image')
plt.imshow(parrot_img, cmap='gray')

plt.subplot(3,3,3)
plt.xticks([])
plt.yticks([])
plt.title('original_image')
plt.imshow(skull_img, cmap='gray')

plt.subplot(3,3,4)
plt.xticks([])
plt.yticks([])
plt.title('Adaptive image 16x16 filter')
plt.imshow(adapt_16_dental, cmap='gray')


plt.subplot(3,3,5)
plt.xticks([])
plt.yticks([])
plt.title('Adaptive image 16x16 filter')
plt.imshow(adapt_16_parrot, cmap='gray')


plt.subplot(3,3,6)
plt.xticks([])
plt.yticks([])
plt.title('Adaptive image 16x16 filter')
plt.imshow(adapt_16_skull, cmap='gray')
```

```python
plt.figure(6, figsize=(14,6))

plt.subplot(3,3,1)
plt.xticks([])
plt.yticks([])
plt.title('original dental image')
plt.hist(dental_img.reshape(-1))
#plt.imshow(np.histogram(skull_img.reshape(-1)), cmap='gray')

plt.subplot(3,3,2)
plt.xticks([])
plt.yticks([])
plt.title('original parrot image')
plt.hist(parrot_img.reshape(-1))
#plt.imshow(parrot_img, cmap='gray')

plt.subplot(3,3,3)
plt.xticks([])
plt.yticks([])
plt.title('original skull image')
plt.hist(skull_img.reshape(-1))
#plt.imshow(skull_img, cmap='gray')

plt.subplot(3,3,4)
plt.xticks([])
plt.yticks([])
plt.title('Adaptive dental image 16x16 filter')
plt.hist(adapt_16_dental.reshape(-1))
#plt.imshow(adapt_16_dental, cmap='gray')


plt.subplot(3,3,5)
plt.xticks([])
plt.yticks([])
plt.title('Adaptive parrot image 16x16 filter ')
plt.hist(adapt_16_parrot.reshape(-1))
#plt.imshow(adapt_16_parrot, cmap='gray')


plt.subplot(3,3,6)
plt.xticks([])
plt.yticks([])
plt.title('Adaptive skull image 16x16 filter ')
plt.hist(adapt_16_skull.reshape(-1))
#plt.imshow(adapt_16_skull, cmap='gray')
```
